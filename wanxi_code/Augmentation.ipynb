```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns

import os
import random

from PIL import Image
import cv2
import albumentations as A
```

```python
ip = []
mp =  []
for img_id in os.listdir('train_images'):
    im_path = os.path.join('train_images',img_id)
    msk_path = os.path.join('train_masks',img_id.replace('jpg','png'))
    ip.append(im_path)
    mp.append(msk_path)
    
```

```python
augV = A.Compose([A.VerticalFlip(p=1)])
augH = A.Compose([A.HorizontalFlip(p=1)])  
augT = A.Compose([A.Transpose(p=1)])
augR = A.Compose([A.RandomRotate90(p=1)])
augG = A.Compose([A.GridDistortion(p=1)])
```

```python
import random
temp = []
for i in range(1000):
    r = random.randint(0,261)
    temp.append(r)
```

```python
def save_images(n_img,n_mask,name):
    img = Image.fromarray(n_img.astype('uint8'))
    mask = Image.fromarray(n_mask.astype('uint8'))
    ip = os.path.join("aug_images",name)
    mp = os.path.join('aug_masks',name)
    img.save(ip+'.png')
    mask.save(mp+'.png')
    
l = len(ip)    
for i in range(l):
    img = Image.open(ip[i])
    mask= Image.open(mp[i])
    img = img.resize((512,512))
    mask = mask.resize((512,512))
    img = np.asarray(img)
    mask = np.asarray(mask)
    
    aug = augH(image=img,mask=mask)
    new_img = aug['image']
    new_mask = aug['mask']
    new_name = ip[i].replace(".","\\").split("\\")[1]+"H"
    save_images(new_img,new_mask,new_name)
    
    aug = augV(image=img,mask=mask)
    new_img = aug['image']
    new_mask = aug['mask']
    new_name = ip[i].replace(".","\\").split("\\")[1]+"V"
    save_images(new_img,new_mask,new_name)
    
    aug = augR(image=img,mask=mask)
    new_img = aug['image']
    new_mask = aug['mask']
    new_name = ip[i].replace(".","\\").split("\\")[1]+"R"
    save_images(new_img,new_mask,new_name)
    
    aug = augT(image=img,mask=mask)
    new_img = aug['image']
    new_mask = aug['mask']
    new_name = ip[i].replace(".","\\").split("\\")[1]+"T"
    save_images(new_img,new_mask,new_name)
    
    aug = augG(image=img,mask=mask)
    new_img = aug['image']
    new_mask = aug['mask']
    new_name = ip[i].replace(".","\\").split("\\")[1]+"G"
    save_images(new_img,new_mask,new_name)
    print(i," finished")
```
